```python
numeric_cols = ['math score','reading score']

minmax_scaler = MinMaxScaler()
df.minmax = df.copy()
df.minmax[numeric_cols] = minmax_scaler.fit_transform(df[numeric_cols])
print("Min-Max Normalised Data:\n",df.minmax[numeric_cols].head())
```

```python
bins = [0,50,70,100]
labels=['low','medium','high']

df['column']=pd.cut(df['math score'], bins=bins , labels=labels)
print(df[['column','math score']])
```

```python
numericCols=['reading score','math score']
minmax_scaler = MinMaxScaler()
df.minmax=df.copy()
df.minmax[numericCols]=minmax_scaler.fit_transform(df[numericCols])
print('MinMax Normalized Data :\n',df.minmax[numericCols].head())
```
